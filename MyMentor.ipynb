```python
def split_data(data_frame, source_col, training_dir, testing_dir, split_size):
    """
    Memisahkan data dari DataFrame menjadi data latih dan data uji
    dengan proporsi yang ditentukan.

    Parameter:
    - data_frame: pandas DataFrame, DataFrame yang berisi data
    - source_col: int, indeks kolom dalam DataFrame yang berisi path sumber file
    - training_dir: string, path ke direktori data latih
    - testing_dir: string, path ke direktori data uji
    - split_size: float, proporsi data uji (0.0 - 1.0)

    Return:
    Tidak ada nilai yang dikembalikan.
    """
    # Menghitung jumlah sampel data uji berdasarkan proporsi
    num_testing = int(len(data_frame) * split_size)

    # Mengacak indeks data
    indices = list(range(len(data_frame)))
    random.shuffle(indices)

    # Memindahkan sampel data ke direktori data latih
    for i in indices[:-num_testing]:
        file_path = data_frame.iloc[i, source_col]
        file_name = os.path.basename(file_path)
        dst = os.path.join(training_dir, file_name)
        print("Copying file:", file_path)
        print("Destination:", dst)
        shutil.copyfile(file_path, dst)

    # Memindahkan sampel data ke direktori data uji
    for i in indices[-num_testing:]:
        file_path = data_frame.iloc[i, source_col]
        file_name = os.path.basename(file_path)
        dst = os.path.join(testing_dir, file_name)
        print("Copying file:", file_path)
        print("Destination:", dst)
        shutil.copyfile(file_path, dst)
```

```python
def get_recommendations_by_domicile(Domicile, cosine_similarities=cosine_similarities1, data=Data):
    # Mendapatkan indeks dari domisili yang cocok
    idx = data[data['Domicile'] == Domicile].index[0]
    
    # Mendapatkan skor kesamaan (similarity) dari semua judul
    sim_scores = list(enumerate(cosine_similarities[idx]))
    
    # Mengurutkan domisili berdasarkan skor kesamaan secara menurun
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Mengambil 5 domisili dengan skor kesamaan tertinggi (kecuali domisili itu sendiri)
    sim_scores = sim_scores[1:6]
    
    # Mengambil indeks dari judul-judul tersebut
    mentor_indices = [i[0] for i in sim_scores]
    
    # Mengembalikan judul-judul rekomendasi
    return data['Domicile'].iloc[mentor_indices]


def get_recommendations_by_Latest_edu(Latest_edu, cosine_similarities=cosine_similarities2, data=Data):
    # Mendapatkan indeks dari id yang cocok
    idx = data[data['Latest Education'] == Latest_edu].index[0]
    
    # Mendapatkan skor kesamaan (similarity) dari semua latest edu
    sim_scores = list(enumerate(cosine_similarities[idx]))
    
    # Mengurutkan latest edu berdasarkan skor kesamaan secara menurun
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Mengambil 5 latest edu dengan skor kesamaan tertinggi (kecuali latest edu itu sendiri)
    sim_scores = sim_scores[1:6]
    
    # Mengambil indeks dari latest edu tersebut
    mentor_indices = [i[0] for i in sim_scores]
    
    # Mengembalikan id-id rekomendasi
    return data['Latest Education'].iloc[mentor_indices]
```

```python
def get_recommendations_by_domicile_new(input_domicile, cosine_similarities=cosine_similarities1, data=Data):
    input_domicile = input_domicile.lower()  # Ubah input ke huruf kecil untuk pembandingan yang tidak peka terhadap besar-kecil huruf
    
    # Filter Data berdasarkan domisili yang sesuai dengan input
    filtered_data = data[data['Domicile'].str.lower() == input_domicile]
    
    if len(filtered_data) == 0:
        return "Tidak ada mentor dengan domisili yang sesuai."
    
    # Hitung similarity score
    idx = filtered_data.index[0]
    sim_scores = list(enumerate(cosine_similarities[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    mentor_indices = [i[0] for i in sim_scores]
    
    return data['Name'].iloc[mentor_indices]
```
